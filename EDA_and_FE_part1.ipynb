```python
# find out all the unique values from col
data_copy["Size"].unique()
```

```python
# converting the unique values into
# meaning full values 
# .str is necessary to access the string accessor methods in pandas.

data_copy["Size"] = data_copy["Size"].str.replace("M" , "000")
```

```python
data_copy["Size"] = data_copy["Size"].str.replace("k" , "")
```

```python
# To remove the strings from the data 
import numpy as np 
data_copy["Size"] = data_copy["Size"].str.replace("Varies with device", str(np.nan))
```

```python
# Making another changes in data for better data cleaning
for i in data_copy["Size"] :
    if i < 10 :
        data_copy["Size"] = data_copy["Size"].replace(i , i*1000)
```
